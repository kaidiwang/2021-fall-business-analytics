# Data Visualization

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

## Crosstabulation

```python
df = pd.read_csv('restaurant.csv')
df
```

```python
# create cross tabulation
pd.crosstab(df['Quality Rating'], df['Meal Price ($)'])
```

```python
# group meal prices into groups and create cross tabulation using the groups of prices
df['price_group'] = pd.cut(df['Meal Price ($)'], bins = [9,19,29,39,49])
df
```

```python
pd.crosstab(df['Quality Rating'], df['price_group'])
```

## scatter chart

```python
import pandas as pd
df = pd.read_csv('electronic.csv')
df
```

```python
# scatter plot
from matplotlib import pyplot as plt
plt.figure(0)
plt.scatter(df['commercials'], df['sales'])
# df.plot.scatter(x='commercials', y='sales')

plt.figure(1)
plt.plot(df['week'], df['sales'], marker = '.')
# df.plot.line(x='week', y='sales')
```

```python
df = pd.read_csv('sales.csv')
# column chart
plt.figure(0)
plt.bar(df['name'], df['sales'])

# bar chart
plt.figure(1)
plt.barh(df['name'], df['sales'])
```

```python
# sort values
df_new = df.sort_values(by = 'sales')
print(df)

# column chart
plt.figure(0)
plt.bar(df_new['name'], df_new['sales'])

# column chart
plt.figure(1)
plt.barh(df_new['name'], df_new['sales'])
```

```python
# pie chart
plt.pie(df_new['sales'], labels = df_new['name'], autopct='%1.1f%%')
plt.show()
```

```python
# bubble chart
df = pd.read_csv('billionaires.csv')
df
```

```python
# plot the bubble plot
plt.scatter(df['Billionaires per 10M Residents'], df['Per Capita Income'], 
            df['Number of Billionaires'])

# label the point using the country names
for i in [0,1,2,3,4,5]:
    plt.text(x=df.loc[i, 'Billionaires per 10M Residents'], 
             y=df.loc[i, 'Per Capita Income'], 
             s=df.loc[i, 'Country'])

# set the x and y axis labels
plt.xlabel('Billionaires per 10M Residents')
plt.ylabel('Per Capita Income')
```
