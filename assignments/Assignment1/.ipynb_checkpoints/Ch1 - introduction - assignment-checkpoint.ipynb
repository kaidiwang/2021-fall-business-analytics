Q5: Impute the population variance of the percent of male (i.e., male/(male+female)). What insights can you get about the distribution of male and female? (20pts)

```python
df['pct_male'] = df['male'] / (df['male'] + df['female'])
df['pct_male'].var(ddof = 0)
df['pct_male'].hist(bins=20)
```

Q6: Assume you want to know about the variance of the low-income households (i.e., income < 50k, corresponding to columns hh_incm_0k, hh_incm_10k, ......, hh_incm_45k). You decide to extract 500 block groups and impute the sample variance to gain insights. How do you achieve the goal using Python and the dataset provided? Please respond with a sample script snippet and the result. Notice that sample variance is different from population variance.(20pts)

```python
cols = ['hh_incm_0k', 'hh_incm_10k', 'hh_incm_15k', 
        'hh_incm_20k', 'hh_incm_25k', 'hh_incm_30k', 
        'hh_incm_35k', 'hh_incm_40k', 'hh_incm_45k']
df_new = df[cols]
# df['hh_incm_0k'] + df['hh_incm_10k'] + .... + df['hh_incm_45k']
df_count = df_new.sum(axis = 'columns')
rand_idx = np.random.choice(1000, size=500,replace = False)
print(rand_idx)
df_low = df_count.iloc[rand_idx]
df_low.var()
```
